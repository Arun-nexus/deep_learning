```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans,DBSCAN
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=load_iris()
x=df.data
scaler=StandardScaler()
x=scaler.fit_transform(x)
print(x)
pca=PCA(n_components=2)
x=pca.fit_transform(x)
print("pca",x)
minmax=MinMaxScaler()
x=minmax.fit_transform(x)
print('minmax',x)
```

```python
inertia=[]
k=range(1,11)
for i in k:
    kmodel=KMeans(n_clusters=i,random_state=0)
    kmodel.fit(x)
    inertia.append(kmodel.inertia_)
```

```python
plt.plot(k,inertia)
plt.grid()
plt.show()
```

```python
model=KMeans(n_clusters=3,random_state=10)
model.fit(x)
y_pred=model.predict(x)
center=model.cluster_centers_
```

```python
plt.scatter(x[:,0],x[:,1],c=y_pred,cmap="Accent")
plt.scatter(center[:,0],center[:,1],marker="x",color="red",s=200,linewidths=2)
plt.show()
```

```python
dbscan=DBSCAN(eps=0.5,min_samples=5)
ypred=dbscan.fit_predict(x)
```

```python
plt.scatter(x[:,0],x[:,1],c=ypred,cmap="Accent")
plt.show()
```
